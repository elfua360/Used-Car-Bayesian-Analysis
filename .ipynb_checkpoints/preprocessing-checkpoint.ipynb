```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
df = pd.read_csv("vehicles-florida.csv")
```

```python
lux = ["lincoln","mercedes-benz","cadillac","lexus","audi","bmw","ferrari","tesla","aston-martin","alfa-romeo","jaguar"]
man = df["manufacturer"]
```

```python
df.head()
```

```python
man = man.array
for i,val in enumerate(man):
    if val in lux:
        man[i] = "luxury"
    else:
        man[i] = "normal"
```

```python
df["manufacturer"] = man
```

```python
df.head(50)
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0', 'region'], axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
df.to_csv("vehicles-final(USE)")
```
