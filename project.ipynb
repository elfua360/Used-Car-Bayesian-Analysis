```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("vehicles.csv")
```

```python
df.head()
```

```python
print(df.columns) #These are all the columns/potential variables, I will drop some that we don't need
```

```python
fl = df[df['state'] == 'fl']
```

```python
fl.head()
```

```python
# Drop the variables we don't need
fl = fl.drop(['id','url','region_url','model','vin', 'image_url', 'description', 'county', 'state','lat','long'], axis=1)
```

```python
fl.head()
```

```python
# Drop specific models of car (There are too many categories)
fl = fl.drop(['model'], axis=1)
```

```python
fl.head()
```

```python
na_counts = []
for i in fl.columns:
    na_counts.append(fl[i].isna().sum())
print('Total rows: ' + str(fl.shape[0]))
print('List NA counts: ')
print(na_counts)
```

```python
# from above the 12th col seems to have a high number of NA values, what is it?
print(fl.columns[11])
```

```python
fl_skim = fl.dropna() # Drop all NAs, we can choose to use a different approach, but this is easier
```
